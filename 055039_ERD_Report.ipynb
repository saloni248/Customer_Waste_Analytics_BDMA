# Analytical Report: Technical Breakdown     

## 1. Database Design & Entity-Relationship Diagram (ERD)  
The ERD visually represents the relationships between entities such as **Customers, Products, Orders, WasteMetrics, Recommendations, Suppliers, Inventory, and EnvironmentalImpactSummary**.  
The primary objective of the ERD is to structure the data effectively to support efficient data retrieval and management.  

### a) Key Design Considerations:  
- Ensured **one-to-many** and **many-to-many** relationships to avoid data redundancy.  
- Optimized data storage with **foreign key constraints** and **referential integrity rules**.  

### b) Key Entities & Relationships  
- **Customers** place orders for **Products**.  
- **Products** are supplied by **Suppliers** and stored in **Inventory**.  
- **Orders** generate **WasteMetrics** that track environmental impact.  
- **Recommendations** are provided to **Customers** based on product usage.  
- **EnvironmentalImpactSummary** consolidates key environmental insights.  




## Relationships Between Tables

Understanding the relationships between different tables is critical in ensuring a robust and efficient database structure. Below are the key relationships between the tables:

* Customers → Orders (One-to-Many): A customer can place multiple orders, but each order is linked to only one customer.

* Products → Orders (One-to-Many): A single product can be part of multiple orders, but each order contains only specific products.

* Orders → WasteMetrics (One-to-One): Each order can have associated waste data, such as return reasons and disposal rates.

* Products → WasteMetrics (One-to-One): Each product is assessed for waste metrics, including recyclability scores and carbon footprint.

* Products → Inventory (One-to-One): Each product has a corresponding stock level, restocking threshold, and storage location.

* Products → EnvironmentalImpactSummary (One-to-One): Each product has an aggregated environmental impact assessment.

* Suppliers → Products (One-to-Many): A supplier can provide multiple products, but each product has only one primary supplier.

* Customers → Recommendations (One-to-Many): Each customer receives multiple recommendations based on their purchasing habits.

## **Data Flow Diagram (DFD) / Process Flow**

A Data Flow Diagram (DFD) represents the movement of data across different components of the system. Unlike an ERD, which focuses on data structure, a DFD shows how data is processed and flows between different entities.

 **Simplified Data Flow**

* **Customer places an order** → Data is stored in the **Orders** table.

* **Product details are fetched** → Data is retrieved from the **Products** table.

* **If the order is returned** → The system updates the **WasteMetrics** table.

* **Waste impact is analyzed** → The **EnvironmentalImpactSummary** table updates sustainability scores.

* **Recommendations are generated** → Customers receive sustainability-based product suggestions from the **Recommendations** table.

This structured flow ensures efficient data management, tracking product performance and waste impact in real time.

## **Overview of the Process Flow**
This Data Flow Diagram (DFD) represents how data moves across different components in an E-commerce Waste Management System. It tracks orders, returns, waste analysis, and sustainable product recommendations to optimize sustainability efforts.

### **Step-by-Step Flow Explanation**

### 1. Customer Interaction
* The Customer initiates the process by placing an order.
The order details are sent to the Order Processing module.

### 2. Order Processing
* The Order Processing module captures order details and stores them in the Orders database.
* It then sends a request to the Product Management system to fetch product details.
The Products database retrieves and supplies the necessary product specifications.

### 3. Handling Returns and Waste Metrics
* If the Customer returns a product, it is processed by the Return Processing module.
* This module updates the WasteMetrics table, which keeps track of returned products and disposal data.

### 4. Sustainability Analysis
* The WasteMetrics table sends data to the Sustainability Analysis module, which updates the EnvironmentalImpactSummary database.
* This step ensures that returned product data contributes to measuring sustainability impact.

###5. Generating Sustainability Recommendations
* Once the sustainability analysis is complete, the Recommendation Generation system generates eco-friendly product suggestions based on waste analysis.
* The Recommendations database stores these insights and sends sustainable product recommendations back to the Customer.

### 6. Supplier Data Integration
* The Supplier provides product and sustainability data, which is integrated into the Products database to ensure accurate tracking of product attributes.

### 7. System Completion
* Once all processes are completed, data is stored in relevant databases (Orders, Products, WasteMetrics, and EnvironmentalImpactSummary).
* The system ensures real-time tracking of waste impact and informs customers about sustainable purchasing options.

### Key Takeaways
* The **DFD effectively maps** the flow of data from customers, orders, returns, and sustainability tracking.
* The **WasteMetrics and EnvironmentalImpactSummary** databases are crucial for tracking waste and sustainability.
* The **Recommendation Generation** system plays a key role in promoting eco-friendly products to customers.
* This structured system **ensures better waste management** while encouraging sustainable shopping habits.
